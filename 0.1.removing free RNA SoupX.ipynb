# Removal of cell free mRNA contamination by SoupX

```r
library(Seurat)
library(SoupX)
```

```r
getwd()
setwd("/media/inspur/AS2150G2/LDK/sclc_metastasis/")
```

```r
samples <- list.files("/media/inspur/AS2150G2/LDK/sclc_metastasis/call ranger outs/", pattern = "^SCLC")
samples
filepath <- paste0("/media/inspur/AS2150G2/LDK/sclc_metastasis/call ranger outs/",samples)
filepath
```

#### ======================================================
#### refer to: https://github.com/constantAmateur/SoupX
**decontaminate free RNA by SoupX using default parameters**
#### ======================================================

```r
# create a list to stash seurat objects
scRNAlist <- list()
```

```r
for (i in 1:length(filepath)) {
  sc <- load10X(filepath[i])
  sc = autoEstCont(sc)
  out = adjustCounts(sc)
  
  # Loading into Seurat
  scRNAlist[[samples[i]]] = CreateSeuratObject(out)
}
```
